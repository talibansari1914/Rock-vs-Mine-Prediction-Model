```python
# Seprating the Data and Labels
X=sonar_data.drop(columns=60 , axis=1)
y=sonar_data[60]

print(X)
print(y)
```

# Training and Splitting The Data


```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,stratify=y,random_state=42)
print(X.shape , X_train.shape , X_test.shape)
```

```python
print(X_train)
print(X_test)
```

```python
print(y_train.shape , y_train)
print(y_test.shape , y_train)
```

# Model

```python
model=LogisticRegression()

# Model Training
model.fit(X_train , y_train)
```

```python
# Prediction & Accuracy On Trainig Data
X_train_predict=model.predict(X_train)
accuracy_train=accuracy_score(y_train,X_train_predict)
print(f"Accuracy Score of the Model on Training Data :{accuracy_train:.4f}")
```

```python
# Prediction & Accuracy On Testing Data
X_test_predict=model.predict(X_test)
accuracy_test=accuracy_score(y_test,X_test_predict)
print(f"Accuracy Score of the Model on Testing Data :{accuracy_test:.4f}")
```
